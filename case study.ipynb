```python
# pip install pygame

import pygame
import random
from enum import Enum
from collections import namedtuple

pygame.init() # khởi tạo

font = pygame.font.SysFont('arial', 25)

class Direction(Enum):
    RIGHT = 1
    LEFT = 2
    UP = 3
    DOWN = 4
    
Point = namedtuple('Point', 'x, y')

# màu rgb
WHITE = (255, 255, 255)
RED = (200,0,0)
BLUE = (0, 0, 255)
BLACK = (0,0,0)

BLOCK_SIZE = 20
SPEED = 6

class SnakeGame:
    
    def __init__(self, w=640, h=480):
        self.w = w
        self.h = h
        # khởi tạo màn hình
        self.display = pygame.display.set_mode((self.w, self.h))
        pygame.display.set_caption('Snake')
        self.clock = pygame.time.Clock() # kiểm soát tốc độ game
        
        # khởi tạo hướng bắt đầu
        self.direction = Direction.RIGHT
        
        # khởi tạo vị trí đầu của con rắn
        self.head = Point(self.w/2, self.h/2)
        
        # khởi tạo con rắn
        self.snake = [self.head, 
                      Point(self.head.x-BLOCK_SIZE, self.head.y),
                      Point(self.head.x-(2*BLOCK_SIZE), self.head.y)]
        # khởi tạo số điểm đầu tiên
        self.score = 0
        # khởi tạo thức ăn và vị trí thức ăn
        self.food = None
        self._place_food() 
    def _place_food(self):
        x = random.randint(0, (self.w-BLOCK_SIZE )//BLOCK_SIZE )*BLOCK_SIZE 
        y = random.randint(0, (self.h-BLOCK_SIZE )//BLOCK_SIZE )*BLOCK_SIZE
        self.food = Point(x, y)
        
    def play_step(self):
        # thu nhập thông tin của người dùng nhập vào
        for event in pygame.event.get():        
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LEFT:
                    self.direction = Direction.LEFT
                elif event.key == pygame.K_RIGHT:
                    self.direction = Direction.RIGHT
                elif event.key == pygame.K_UP:
                    self.direction = Direction.UP
                elif event.key == pygame.K_DOWN:
                    self.direction = Direction.DOWN
        
        # 2. di chuyển
        self._move(self.direction) # cập nhật đầu
        self.snake.insert(0, self.head)
        
        # kiểm tra xem trò chơi kết thúc chưa
        game_over = False
        if self._is_collision():
            game_over = True
            return game_over, self.score
            
        # đặt lại thức ăn ở vị trí mới
        if self.head == self.food:
            self.score += 1 # tăng thêm số điểm 
            self._place_food()
        else:
            self.snake.pop()
        
        # cập nhật lại giao diện
        self._update_ui()
        self.clock.tick(SPEED)
        
        # thoát trò chơi và in ra số điểm
        return game_over, self.score
    
    # hàm kiểm tra va chạm
    def _is_collision(self):
        # va vào thành
        if self.head.x > self.w - BLOCK_SIZE or self.head.x < 0 or self.head.y > self.h - BLOCK_SIZE or self.head.y < 0:
            return True
        # va vào chính nó
        if self.head in self.snake[1:]:
            return True
        
        return False
        
    def _update_ui(self):
        self.display.fill(BLACK)
        
        for pt in self.snake:
            pygame.draw.rect(self.display, BLUE, pygame.Rect(pt.x, pt.y, BLOCK_SIZE, BLOCK_SIZE))
            
        pygame.draw.rect(self.display, RED, pygame.Rect(self.food.x, self.food.y, BLOCK_SIZE, BLOCK_SIZE))
        
        text = font.render("Score: " + str(self.score), True, WHITE)
        self.display.blit(text, [0, 0])
        pygame.display.flip()
        
    # hàm điểm khiển con rắn
    def _move(self, direction):
        x = self.head.x
        y = self.head.y
        if direction == Direction.RIGHT:
            x += BLOCK_SIZE
        elif direction == Direction.LEFT:
            x -= BLOCK_SIZE
        elif direction == Direction.DOWN:
            y += BLOCK_SIZE
        elif direction == Direction.UP:
            y -= BLOCK_SIZE
            
        self.head = Point(x, y)
            

if __name__ == '__main__':
    game = SnakeGame()
    
    # vòng lặp game
    while True:
        game_over, score = game.play_step()
        
        if game_over == True:
            break
        
    print('Final Score', score)
        
        
    pygame.quit() # thoát trò chơi
```
